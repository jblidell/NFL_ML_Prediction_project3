```python
import pandas as pd
from pathlib import Path
```

```python
off_stat_df = pd.read_csv(Path('./Resources/off_stat_2122_rbsdm.csv'))

display(off_stat_df.head())
display(off_stat_df.tail())
```

```python
off_stat_df = off_stat_df.drop(['Unnamed: 0','Team','Success Rate (SR)','Dropback EPA','Dropback SR','Rush EPA','Rush SR'],axis=1)
off_stat_df = off_stat_df.sort_values(by=['Abbr'])
off_stat_df
```

```python
off_stat_df = off_stat_df.rename({'EPA/play':'O_EPA/play'}, axis='columns')
off_stat_df
```

```python
def_stat_df = pd.read_csv(Path('./Resources/def_stat_2122_rbsdm.csv'))

display(def_stat_df.head())
display(def_stat_df.tail())
```

```python
def_stat_df = def_stat_df.drop(['Unnamed: 0','Team','Success Rate (SR)','Dropback EPA','Dropback SR','Rush EPA','Rush SR'],axis=1)
def_stat_df
```

```python
def_stat_df = def_stat_df.rename({'EPA/play':'D_EPA/play'}, axis='columns')
def_stat_df = def_stat_df.sort_values(by=['Abbr'])
def_stat_df
```

```python
team_O_D_EPA = off_stat_df.merge(def_stat_df, left_on='Abbr', right_on='Abbr')
team_O_D_EPA
```
